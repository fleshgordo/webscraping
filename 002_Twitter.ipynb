### Erster Tweet

Mit der Funktion update_status() könnt ihr einen Tweet an den Twitter Account senden. Achtung: in den Klammern müssen zwei Anführungsstriche sein. Bitte ändert den Text bevor ihr das Feld ausführen lässt:

### Such-Funktion

Man kann nach Schlagwörter suchen und bekommt eine Liste mit Tweets zurück. Die Funktion hierfür heisst "search". Der Suchbegriff muss in den Klammern stehen api.search("suchwort")

```python
api.search("trump")
```

Da die Liste etwas lang und unübersichtlich ist, speichere ich sie zunächst in einer Variablen ab:

```python
results = api.search("sarahbland")
```

Hat das geklappt? Ich kann einfach die print Funktion aufrufen und sehen ob was im Terminal erscheint:

```python
print results
```

Jetzt wird die for-Schleife wieder interessiert, weil wir diese verwenden können, um die einzelnen Einträge anzuzeigen:

```python
for tweet in results:
    print tweet
```

Das sieht nach wie vor recht unübersichtlich aus. Aber man kann nun einzelne Daten einfach mit der Punkt.Syntax anzeigen lassen. Probiert mal einen Punkt nach tweet. einzugeben und danach die Tabulator Taste:

```python
for tweet in results:
    print tweet.text
```

Wenn wir uns die Tweepy Hilfe ansehen, dann kann die Suchfunktion noch erweitert werden

API.search(q[, lang][, locale][, rpp][, page][, since_id][, geocode][, show_user])

Returns tweets that match a specified query.

Parameters:	
    - q – the search query string
    - lang – Restricts tweets to the given language, given by an ISO 639-1 code.
    - locale – Specify the language of the query you are sending. This is intended for language-specific clients and the default should work in the majority of cases.
    - rpp – The number of tweets to return per page, up to a max of 100.
    - page – The page number (starting at 1) to return, up to a max of roughly 1500 results (based on rpp * page.
    - since_id – Returns only statuses with an ID greater than (that is, more recent than) the specified ID.
    - geocode – Returns tweets by users located within a given radius of the given latitude/longitude. The location is preferentially taking from the Geotagging API, but will fall back to their Twitter profile. The parameter value is specified by “latitide,longitude,radius”, where radius units must be specified as either “mi” (miles) or “km” (kilometers). Note that you cannot use the near operator via the API to geocode arbitrary locations; however you can use this geocode parameter to search near geocodes directly.
    - show_user – When true, prepends “<user>:” to the beginning of the tweet. This is useful for readers that do not display Atom’s author field. The default is false.
    
Wie kann man z.B. Tweets nur in Zurich suchen

```python
api.search(q="trump",geocode="47.3673,8.55,10km")
```

```python
results = api.search(q="trump",geocode="47.3673,8.55,10km")

for tweet in results:
    print tweet.text
```

### In ein File schreiben

Wir wollen diesen Text nun in ein PDF verwandeln. Wie gehen wir hier vor? Zunächst müssen wir die Informationen zwischenspeichern. Am einfachsten ist es, den Text in ein Textfile zu schreiben. In Python kann man ein File neu erstellen, Zeile für Zeile etwas hinzufügen und danach das File abspeichern. Das geht mit dem Befehl open():

```python
f=open('tweets.md','w')
```

```python
f.write('hello world')
f.close()
```

Mit dem Befehl write("") fügt man Text dem File hinzu und mit close() schliesst man den "Text"-Handler. Schaut nun nach, ob das File tatsächlich im Ordner liegt
